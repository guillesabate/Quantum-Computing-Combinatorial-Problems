## Equal Superposition Layer

```python
import matplotlib as mpl
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter, ParameterVector
mpl.rcParams['figure.dpi'] = 300

size = 2
circuit = QuantumCircuit(size)
circuit.h(range(size))

circuit.draw("mpl")
```

## Cost Layer

```python
size = 2
circuit = QuantumCircuit(size)

parameter = Parameter('γ[0]')

circuit.cx(0,1)
circuit.rz(parameter, 0)
circuit.cx(0,1)

circuit.draw("mpl")
```

## Mixer Layer

```python
size = 2
circuit = QuantumCircuit(size)

parameter = Parameter('θ[0]')

circuit.rx(parameter, range(size))

circuit.draw("mpl")
```

### Measurement Layer

```python
size = 2
circuit = QuantumCircuit(size, size)

for i in range(size):
    circuit.measure(i, i)

circuit.draw("mpl")
```

## Actual QAOA Circuit

```python
size = 2
layers = 1
circuit = QuantumCircuit(size, size)
parameter_gamma = Parameter('γ[0]')
parameters_beta = ParameterVector('β', size*layers)


circuit.h(range(size))

circuit.barrier(range(size))

circuit.cx(0,1)
circuit.rz(parameter_gamma, 0)
circuit.cx(0,1)

for i in range(size*layers):
    circuit.rx(parameters_beta[i], i)

circuit.barrier(range(size))

for i in range(size):
    circuit.measure(i, i)

circuit.draw("mpl")
```
